```python
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import tensorflow as tf

questions = []
answers = []

for question, answer in data:
  for z in['.',',','?','!']:
    question = question.replace(z,"")
    answer = answer.replace(z,"")
  questions.append(question.strip().lower())
  answers.append(answer.strip().lower())

print(questions,'\n',answers)
```

```python
tokenizer_q = Tokenizer()
tokenizer_q.fit_on_texts(questions)
tokenizer_a = Tokenizer()
tokenizer_a.fit_on_texts(answers)

sequences_q = tokenizer_q.texts_to_sequences(questions)
sequences_a = tokenizer_a.texts_to_sequences(answers)

max_length_q = int(max(len(seq) for seq in sequences_q))
max_length_a = int(max(len(seq) for seq in sequences_a))

X = pad_sequences(sequences_q, maxlen=max_length_q)
y = pad_sequences(sequences_a, maxlen=max_length_a)
print(X,'\n',y)
#
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Embedding, TimeDistributed

vocab_size_q = len(tokenizer_q.word_index)+1
vocab_size_a = len(tokenizer_a.word_index)+1

model = Sequential()
model.add(Embedding(input_dim=vocab_size_q, output_dim=64, input_length=max_length_q))
model.add(tf.keras.layers.Bidirectional(LSTM(64, return_sequences=False)))
model.add(tf.keras.layers.Dropout(0.1))
model.add(tf.keras.layers.RepeatVector(max_length_a))
model.add(tf.keras.layers.Bidirectional(LSTM(64, return_sequences=True)))
model.add(tf.keras.layers.BatchNormalization())
model.add(TimeDistributed(Dense(vocab_size_a, activation='softmax')))
model.add(tf.keras.layers.Dropout(0.1))
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam',metrics=['accuracy'])

model.summary()
#
```

```python
model.fit(X_train, np.expand_dims(y_train, -1),
          validation_data=(X_test, y_test), epochs=120)
#
```
